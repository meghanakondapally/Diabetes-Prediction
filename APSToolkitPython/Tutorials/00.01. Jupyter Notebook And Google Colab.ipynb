## Google Colab

Google Colaboratory, often referred to as "Google Colab," is a cloud-based Python development environment similar to Jupyter Notebook. It allows you to write and execute Python in your browser, with zero configuration required, free access to GPUs, and easy sharing of your work.

Here's a simple example of how to use Google Colab:

1. First, you need to access the Google Colab website (https://colab.research.google.com) and sign in with your Google account.

2. Click on `File > New notebook` to create a new notebook.

3. You'll see a cell where you can start writing your Python code. For example, let's write a simple Python program that prints "Hello, World!".



````python
print("Hello, World!")
````



4. To run the code, click on the play button on the left side of the cell or press `Shift + Enter`. The output will be displayed below the cell.

5. You can add new cells by clicking the `+ Code` or `+ Text` buttons. `+ Code` adds a new cell for Python code, while `+ Text` adds a new cell for markdown text.

6. Once you're done, you can save your notebook to Google Drive or download it to your local machine by clicking on `File > Save` or `File > Download .ipynb`.

Remember, Google Colab also supports free GPU usage. You can change your runtime to use GPU by clicking on `Runtime > Change runtime type > Hardware accelerator > GPU`.
